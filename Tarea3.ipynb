```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
DB = pd.read_csv('Churn_Modelling.csv')
(n,d)=np.unique(DB['Geography'],return_counts=True)
print(n)
```

```python
X = DB.iloc[:, 3:13].values
Y = DB.iloc[:, 13].values
print(X)
print(Y)
print(X.shape)

```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

# Country column
ct = ColumnTransformer(transformers=[('cat', OneHotEncoder(), [1, 2])], remainder='passthrough')

X = ct.fit_transform(X)

print(X[0])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=69)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1875, random_state=69)

print(X_train.shape)
print(X_test.shape)
print(X_val.shape)
```
